```python
%matplotlib inline
from matplotlib import pyplot as plt
from scurve_lib import *

# Files
sc_file_dark = "data/sc_dark_ECg_4aug.dat"
sc_file_light = "data/sc_flatsc_HV3900_ECg_4aug.dat"
```

```python
# Dark (HV DAC 3900)
sc_avg_dark = read_scurve(sc_file_dark)
```

```python
# Plot 
plt.figure(figsize=(20,7))
plt.plot(sc_avg_dark, color = 'k', alpha = 0.2, linewidth = 3)
plt.xlabel('DAC')
plt.title('Dark Data')
plt.ylabel('#')
plt.savefig('plots/dark_data.png')
plt.show()
```

```python
# Flatscreen (HVDAC 3900)
sc_avg_light = read_scurve(sc_file_light)
```

```python
# Plot 
plt.figure(figsize=(20,7))
plt.plot(sc_avg_light, color = 'k', alpha = 0.2, linewidth = 3)
plt.xlabel('DAC')
plt.title('Light Data')
plt.ylabel('#')
plt.savefig('plots/light_data.png')
plt.show()
```

```python
# Plot single
dac = range(1000)
fig = plt.figure(figsize=(20,7))
print np.shape(sc_avg_light)
tp = np.transpose(np.array(sc_avg_light))
plt.plot(dac, tp[0][:], dac, tp[999][:], dac, tp[1000][:], alpha = 0.9, linewidth = 3)
plt.ylim(0, 5)
plt.show()

# average over 10 Levels..


```
